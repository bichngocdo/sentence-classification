```python
import pandas as pd

from sentclf.experiment import Experiment
```

# Data

## Load Data

```python
data_path = '../SPSC.csv'

data = pd.read_csv(data_path)
data['sentence'] = data['sentence'].astype('str')
data['label'] = data['label'].astype('category')
data.head()
```

# Experiment

## Config

```python
config = {
    'data': {
        'random_states': [
              98237834,
              124104,
        ],
    },
    'dataset': {
        'name': 'Uizard',
    },
    'data_loaders': {
        'train_batch_size': 32,
        'dev_batch_size': 64,
        'test_batch_size': 64,
    },
    'model': {
        'pretrained_model': 'allenai/scibert_scivocab_uncased',
        'pooling_mode': 'mean',
        'dropout_prob': 0.1,
        'use_extended_features': False,
    },
    'optimizer': {
        'type': 'Adam',
        'learning_rate': 5e-6,
        'learning_rate_top': 5e-4,
    },
    'trainer': {
        'num_epochs': 1,
        'interval': 200,
    },
    'paths': {
        'output_dir': 'model'
    },
    'logger': {
        'level': 'INFO',
    }
}
```

## Initialization

```python
experiment = Experiment(
    config,
    data=data,
)
```

```python
experiment.train_data.info()
```

```python
experiment.dev_data.info()
```

## Training

```python
experiment.train()
```

## Test Results

```python
experiment.trainer.load_best_model()
experiment.test()
```
